```python
import pandas as pd
```

```python
df_crime = pd.read_csv("censusCrimeClean.csv")

income = df_crime["medIncome"] 
violent_crimes = df_crime["ViolentCrimesPerPop"]


```

```python
from scipy.stats import stats
```

```python
correlation = stats.pearsonr(income,violent_crimes)
print(correlation)
```

```python
import matplotlib.pyplot as plt
plt.suptitle('medIncome vs. ViolentCrimesPerPop')
plt.xlabel('medIncome')
plt.ylabel('ViolentCrimesPerPop')
plt.scatter(income,violent_crimes , c = "#D06B36", s = 10, alpha = 0.4, linewidth='0')

plt.show()
```

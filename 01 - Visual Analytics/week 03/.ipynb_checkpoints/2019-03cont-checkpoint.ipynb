```python
# Importing Libraries
import os

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import pyplot
from mpl_toolkits.mplot3d import Axes3D

from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

import folium

from collections import Counter

np.random.seed(123)

# SETTING THE WORKING DIRECTORY : Change the path to where your data is
#os.chdir('.')
```

```python
# Loading csv file into Pandas DataFrame
df = pd.read_csv('storm_tweets.csv')
#df.head()
```

```python
# Transforming the type of MESSAGEDATE into datetime
df['MESSAGEDATE'] = pd.to_datetime(df['MESSAGEDATE'], format='%d/%m/%Y %H:%M:%S')
print('The column MESSAGEDATE is of type :', df.MESSAGEDATE.dtype)
print('The min value of MESSAGEDATE is', df.MESSAGEDATE.min(), 'whereas the max is :', df.MESSAGEDATE.max())
```

```python
# In order to inspect temporal distribution and aggregate the time attribute flexibly,
# it's interesting to set the MESSAGEDATE column as an index of the dataframe, this will allow us to use the 
# powerful resample() method !
df.set_index(df['MESSAGEDATE'], inplace=True)
print('The index type of the dataframe is', type(df.index), '\n')
#df.head()
```

```python
# Inspecting the temporal distribution by DAY
df_count_daily = df.resample('D').count()[['MESSAGETEXT']]

# Visualize the evolution
plt.figure(figsize = (15, 5))
plt.xlabel('Date by Day')
plt.ylabel('Number of Tweets')
plt.title('Daily evolution of the number of tweets about the storm')
plt.bar(df_count_daily.index, df_count_daily.values.flatten());
```

<b>Commentary :</b>

<p>
    As expected, there is an upward trend of the number of tweets about the storm between 2015-03-27 and 2015-03-31 with a sudden jump on the 31th before declining the day afterwards. Since most twitter users are obsessive maniacs who'd instinctively tweet about the fire in their house before actually calling the fire and rescue service (no offense lol), we can safely assume that the number of tweets represents an instant reaction to what's happening in the same day and/or moment (i.e there is no big delay between the time of an event and the tweets that follow it). This leads me to believe that the storm hit its peak on the 31th of March.
</p>
<p>
    There is one more important thing to mention here. When dealing with counts, especially comparison of counts, one needs to take into account the total population on which the count was performed so that the values are normalized and can be safely compared. In this case, we can assume that the total number of twitter users is relatively stable during the span of the period, so the counts can be compared and the increase suggests that there is an actual factor affecting the number of tweets : the storm probably.
</p>

<p>
    One last thing ... We need to take the last value (2015-04-02) with a grain of salt : if you remember from the code above, the start date of the tweets was 2015-03-27 00:05:49 while the last tweet was recorded on 2015-04-02 at 13:40:59. This means that we don't have all the tweets for the last day (about 10 hours missing), so the displayed value is probably underestimating the actual number of tweets for that day.
</p>

```python
# Let's refine the analysis by inspecting the temporal distribution by HOUR
df_count_hourly = df.resample('H').count()[['MESSAGETEXT']]

# Display the peak of the hourly count
print('The highest value of the hourly count was', df_count_hourly.MESSAGETEXT.max(), 'and was reached on', df_count_hourly.MESSAGETEXT.idxmax())

# Visualize the evolution
plt.figure(figsize = (20, 5))
plt.xlabel('Date By Hour')
plt.ylabel('Number of Tweets')
plt.title('Hourly evolution of the number of tweets about the storm')
plt.bar(df_count_hourly.index, df_count_hourly.values.flatten(), width = 0.02);
```

<b>Commentary :</b>

<p>
    Moving from Daily count to Hourly count allows us to inspect more closely what happened. The first thing to notice is the hourly seasonality of the data (seasonality is the presence of variations that occur at specific regular intervals - Dr. Wikipedia at your service). Each day, the number of tweets starts increasing from midnight to hit a peak around mid-day before it drops back again all the way until it hits a low around midnight of the next day.
</p>

<p>
    We saw before that the highest daily number of tweets was recorded during the 31th. However, if we inspect the data more closely, we see that the number of tweets on the 30th actually spiked at night instead of the expected decrease, recording the highest hourly value of the entire period : 72 from 23:00 to 00:00. This suggests that the storm started on the night of the 30th and lasted through the 31th before things went back to normal the night of the same day.
</p>

<p>
    One last thing ... again :). I am making hasty conclusions here because I have background information about what actually happened. If this was some other time series I had no prior information about, I wouldn't have jumped to conclusions this fast given the data at hand, and the reason is .... seasonality ! Indeed, if we want to confirm that the behavior on the 30th and 31th is actually due to some external factor affecting the series, we need to inspect previous weeks and make sure there isn't a similar behavior on the same days of the past weeks (in other words, a weekly seasonality that resembles the data we have). Let me give you an example to better understand this concept. If we were to show a time series of the number of daily phone calls for 7 days, from Monday to Sunday, then you would notice that the number is somewhat high from Monday to Friday, and suddenly drops on Saturday and Sunday. Does this mean that something happened on that particular Saturday and Sunday ? Probably not, because if you look at past weeks, the behavior is the same, during week ends, people call less often compared to working days : The time series has a weekly seasonality, which represents its normal behavior.
</p>

<p>
    By the way, the 30th here is a Monday so it's very unlikely that the strange increase is simply due to weekly seasonality.
</p>

<h2>Task III :</h2>

<h2>Explore and characterize the joint distribution of the different keywords. Which combinations of
weather phenomena and/or events that happened during the storm occur frequently in the
tweets? Which are rare? Which keywords tend to occur more in isolation from others and which
in combinations with others?</h2>

```python
# Listing the columns of the dataset
df.columns

# Listing keywords
keywords = ['keyword=storm', 'keyword=wind', 'keyword=rain', 'keyword=snow', 'keyword=hail', 'keyword=flood',
            'keyword=road', 'keyword=collapse', 'keyword=ingury', 'keyword=disrupt', 'keyword=forecast']

# Listing Classes of tweets
tweet_classes = ['bad weather (sum)', 'consequence (sum)', 'bad weather', 'consequence']
```

<b>The first thing we're going to do is analyze the keywords separately</b>

```python
# Visualizing the total occurence of the keywords
plt.figure(figsize=(20, 5))
plt.xlabel('Keywords')
plt.ylabel('Tweets count')
plt.title('Total number of tweets containing each keyword')
plt.bar(keywords, df[keywords].sum());
```

<b>Commentary :</b>

<p>
    The most frequent keywords overall are 'storm', 'wind', 'rain' and 'flood' with 'storm' and 'wind' being far ahead of the rest. This is going to be important later on when we inspect the pairwise combinations of keywords.
</p>

```python
# Let's run a Principal Component Analysis with 2 components on the Keywords alone
pca_1 = PCA(n_components = 2)
pca_1.fit(df[keywords])
df_pca_1 = pca_1.transform(df[keywords])

# Inspect the Variance explained by each Principal Component (to have an idea about the loss of information incurred by the projection)
print('Explained variance ratio \n   Total :', round(pca_1.explained_variance_ratio_.sum()*100, 2), '%',
      '\n   By principal component :', pca_1.explained_variance_ratio_, '\n');
```

<p>
    Note that the total variance explained by our PCA algorithm is ~55%, this means that this projection made us lose almost half of the information in our dataset. This is something you need to keep track of when using PCA, because if you lose too much information, you should adapt your conclusions accordingly.
</p>

```python
# Now let's try to cluster the data using kmeans before visualizing
n_clusters_1 = 4 # Change this value to experiment with different numbers of clusters
kmeans_1 = KMeans(n_clusters=n_clusters_1)

# Fitting the algorithm to the data
kmeans_1 = kmeans_1.fit(df[keywords])

# Getting the Cluster labels for the dataframe
labels_1 = kmeans_1.predict(df[keywords])

# Getting the Centroid values
centroids_1 = kmeans_1.cluster_centers_
centroids_pca_1 = pca_1.transform(centroids_1)
```

```python
# Color dictionary for to encode the clusters
cluster_colors = {0 : 'mediumslateblue', 1 : 'seagreen', 2 : 'tomato', 3 : 'goldenrod', 4 : 'palevioletred'}

# Let us visualize the result in a scatter plot
plt.figure(figsize=(10,8))
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('Projection into the first two Principal Components + Clustering with k = {}'.format(n_clusters_1))

label_colors_1 = list(map(cluster_colors.get, labels_1))
centroid_colors = ['indigo', 'darkgreen', 'indianred', 'saddlebrown']

plt.scatter(df_pca_1[:,0], df_pca_1[:,1], alpha = .2, c = label_colors_1)
plt.scatter(centroids_pca_1[:,0], centroids_pca_1[:,1], marker='s', edgecolors = centroid_colors, linewidth = 3, facecolors = 'None', s=200);
```

```python
# Now let's try the same thing but with the Keywords + "bad weather" + "consequence"
pca_2 = PCA(n_components = 2)
pca_2.fit(df[keywords + ["bad weather", "consequence"]])
df_pca_2 = pca_2.transform(df[keywords + ["bad weather", "consequence"]])

# Inspect the Variance explained by each Principal Component (to have an idea about the loss of information incurred by the projection)
print('Explained variance ratio \n   Total :', round(pca_2.explained_variance_ratio_.sum()*100, 2), '%',
      '\n   By principal component :', pca_2.explained_variance_ratio_, '\n');
```

```python
# Again let's try to cluster the data using kmeans before visualizing
n_clusters_2 = 4 # Change this value to experiment with different numbers of clusters
kmeans_2 = KMeans(n_clusters=n_clusters_2)

# Fitting the algorithm to the data
kmeans_2 = kmeans_2.fit(df[keywords + ["bad weather", "consequence"]])

# Getting the Cluster labels for the dataframe
labels_2 = kmeans_2.predict(df[keywords + ["bad weather", "consequence"]])

# Getting the Centroid values
centroids_2 = kmeans_2.cluster_centers_
centroids_pca_2 = pca_2.transform(centroids_2)
```

```python
# Color dictionary for to encode the clusters
cluster_colors = {0 : 'mediumslateblue', 1 : 'seagreen', 2 : 'tomato', 3 : 'goldenrod', 4 : 'palevioletred'}

# Let us visualize the result in a scatter plot
plt.figure(figsize=(10,8))
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('Projection into the first two Principal Components + Clustering with k = {}'.format(n_clusters_2))

label_colors_2 = list(map(cluster_colors.get, labels_2))
centroid_colors = ['indigo', 'darkgreen', 'indianred', 'saddlebrown']

plt.scatter(df_pca_2[:,0], df_pca_2[:,1], alpha = .2, c = label_colors_2)
plt.scatter(centroids_pca_2[:,0], centroids_pca_2[:,1], marker='s', edgecolors = centroid_colors, linewidth = 3, facecolors = 'None', s=200);
```

<p>
    K-means is a pretty good clustering algorithm overall. It operates by clustering the data into hyperspheres in the original space using some distance metric, Eucliedean distance by default. We note here that the projection from the original space into the new one has an effect of spatial distortion, and so the original distances between data points are no longer the same in the PCA space. Notice that some purple points in the scatter plot above are close to the green centroid and yet they belong to the purple cluster whose centroid is far away. In the original space those points are actually closer to the purple centroid than any other centroid.
</p>

```python
# Cluster Analysis : Number of tweets per Cluster
cluster_names = {0 : 'Cluster 1', 1 : 'Cluster 2', 2 : 'Cluster 3', 3 : 'Cluster 4'}
counter = Counter(labels_2)
plt.figure(figsize = (10, 5))
plt.xlabel('Clusters')
plt.ylabel('Number of tweets')
plt.title('Number of tweets per Cluster')
plt.bar(list(map(cluster_names.get, list(counter.keys()))), list(counter.values()));
```

<p>
    The distribution of tweets per clusters is not too bad, each cluster contains enough tweets to be considered meaningful. This is important for when we're going to inspect the distribution of keywords per cluster, because if one of the clusters had most of the tweets, the distribution of keywords could be biased towards the bigger cluster.
</p>

```python
# Cluster Analysis : Evolution of Clusters per day
df['Cluster'] = labels_2

c1 = df.Cluster[df.Cluster == 0].resample('D').count()
c2 = df.Cluster[df.Cluster == 1].resample('D').count()
c3 = df.Cluster[df.Cluster == 2].resample('D').count()
c4 = df.Cluster[df.Cluster == 3].resample('D').count()

plt.figure(figsize=(15, 5))

p1 = plt.bar(c1.index, c1.values, color = centroid_colors[0])
p2 = plt.bar(c2.index, c2.values, bottom = c1.values, color = centroid_colors[1])
p3 = plt.bar(c3.index, c3.values, bottom = (c1.values + c2.values), color = centroid_colors[2])
p4 = plt.bar(c4.index, c4.values, bottom = (c1.values + c2.values + c3.values), color = centroid_colors[3])

plt.xlabel('Date')
plt.ylabel('Number of tweets')
plt.title('Evolution of the number of tweets by Cluster')
plt.legend((p1[0], p2[0], p3[0], p4[0]), ('Cluster 1', 'Cluster 2', 'Cluster 3', 'Cluster 4'));
```

<p>
    From the stacked bar plot above, we note that the proportions of tweets in each cluster are somewhat constant throughout the entire period. This potentially means that our clusters are stable in time.
</p>

```python
# Cluster Analysis : Distribution of keywords in each cluster


# Visualizing the total occurence of the keywords
plt.figure(figsize=(20, 30))

plt.subplot(4, 1, 1)
plt.xlabel('Keywords')
plt.ylabel('Tweets count')
plt.title('Cluster 1 : Total number of tweets containing each keyword')
plt.bar(keywords, df.loc[labels_2 == 0 ,keywords].sum())

plt.subplot(4, 1, 2)
plt.xlabel('Keywords')
plt.ylabel('Tweets count')
plt.title('Cluster 2 : Total number of tweets containing each keyword')
plt.bar(keywords, df.loc[labels_2 == 1 ,keywords].sum())

plt.subplot(4, 1, 3)
plt.xlabel('Keywords')
plt.ylabel('Tweets count')
plt.title('Cluster 3 : Total number of tweets containing each keyword')
plt.bar(keywords, df.loc[labels_2 == 2 ,keywords].sum())

plt.subplot(4, 1, 4)
plt.xlabel('Keywords')
plt.ylabel('Tweets count')
plt.title('Cluster 4 : Total number of tweets containing each keyword')
plt.bar(keywords, df.loc[labels_2 == 3 ,keywords].sum());
```

<p>
    Now this is more interesting. Looking at the distribution of the keyword in each cluster shows a clear separation, with cluster 1 tweets having keywords all related to weather phenomena as well as weather forecasts and cluster 3 focusing on the consequences like injury, flood and disruption. Cluster 2 groups the tweets mentioning the storm alone without any other keyword from the pre-selected set of keywords. Keep in mind that this is also the cluster with the highest number of tweets. It is reasonable to inspect it more closely, in particular, to look what other words were frequent in the texts. Finally, cluster 4 is heavily focused on the wind and the road.
</p>
